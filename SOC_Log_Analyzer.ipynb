```python
log_data = """
Jul 10 10:12:01 server sshd[1001]: Failed password for invalid user admin from 192.168.1.50 port 22 ssh2
Jul 10 10:12:05 server sshd[1002]: Failed password for invalid user root from 192.168.1.50 port 22 ssh2
Jul 10 10:12:10 server sshd[1003]: Failed password for invalid user test from 192.168.1.50 port 22 ssh2
Jul 10 10:12:15 server sshd[1004]: Failed password for invalid user admin from 192.168.1.50 port 22 ssh2
Jul 10 10:12:20 server sshd[1005]: Failed password for invalid user root from 192.168.1.50 port 22 ssh2
Jul 10 10:13:10 server sshd[1006]: Failed password for invalid user user from 10.0.0.5 port 22 ssh2
"""

with open("auth.log", "w") as file:
    file.write(log_data)

print("Log file created")
```

```python
import re
from collections import defaultdict
```

```python
failed_logins = defaultdict(int)

def analyze_log(file_path):

    with open(file_path, "r") as file:
        logs = file.readlines()

    print("\n----- SOC Log Analysis Report -----\n")

    for line in logs:

        if "Failed password" in line:

            ip_match = re.search(r'\d+\.\d+\.\d+\.\d+', line)

            if ip_match:
                ip = ip_match.group()
                failed_logins[ip] += 1

    for ip, count in failed_logins.items():

        if count >= 5:
            print(f"ALERT: Possible brute force attack from {ip}")
            print(f"Failed login attempts: {count}\n")

        else:
            print(f"Suspicious activity from {ip} ({count} failed attempts)\n")

    print("Analysis Completed")
```

```python
analyze_log("auth.log")
```
